```python
from pandas import DataFrame
import pandas as pd

# 첫번째 데이터프레임
data = {
    '종가': [113000, 111500],
    '거래량': [555850, 282163]
}
index = ["2019-06-21", "2019-06-20"]
df1 = DataFrame(data, index=index)
df1
```

```python
# 두번째 데이터프레임
data = {
    '종가': [110000, 483689],
    '거래량': [109000, 791946]
}
index = ["2019-06-19", "2019-06-18"]

df2 = DataFrame(data, index=index)
df2
```

```python
result = pd.concat([df1,df2])
result
#날짜가 약간 안 맞으니... 순서대로 만들어봅다다
```

```python
from pandas import DataFrame
import pandas as pd

# 첫 번째 데이터프레임
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "051910", "LG화학", 896000],
    ["전기전자", "000660", "SK하이닉스", 101500]
]

columns = ["업종", "종목코드", "종목명", "현재가"]
df1 = DataFrame(data=data, columns=columns)
df1
```

```python
from pandas import DataFrame
import pandas as pd

# 첫 번째 데이터프레임
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "051910", "LG화학", 896000],
    ["서비스업", "035720", "카카오", 101500]
]

columns = ["업종", "종목코드", "종목명", "현재가"]
df1 = DataFrame(data=data, columns=columns)
df1
```

```python
# 두 번째 데이터프레임
data = [
    ["은행", 2.92],
    ["보험", 0.37],
    ["화학", 0.06],
    ["전기전자", -2.43]
]

columns = ["업종", "등락률"]
df2 = DataFrame(data=data, columns=columns)
df2
```

```python
df=pd.merge(left=df1,right=df2,how='left',on='업종')
df
# how가 left라서 일단 left에 있는 것은 전부 나오게 함
# 근데 서비스업의 경우 등락률이 없기 때문에 일단 NAN으로 나옴
```

```python
df=pd.merge(left=df1,right=df2,on='업종')
df
# how가 없다면 inner join이 되어 버려서 left 와 right 모두에 있는 것만 살아 남음
# 양쪽의 키 값이 매칭되는 것만 가져옴
```

```python
df=pd.merge(left=df1,right=df2,how='right',on='업종')
df
# how가 right라서 일단 right에 있는 것은 전부 나오게 함
```

```python
df=pd.merge(left=df1,right=df2,how='outer',on='업종')
df
# outers 로 하면 그냥 전부 나오게 함
# 근데 현실 세계에서 full join은 보통 거의 쓸 일 거의 없음


#merge 사용 시에는 inner 혹은 left 위주로 사용하게 됨
#기준 테이블을 무엇으로 잡느냐의 문제인데 굳이 right로 할 필요는 없음
```

```python
# 첫 번째 데이터프레임
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "051910", "LG화학", 896000],
    ["서비스업", "035720", "카카오", 101500]
]

columns = ["업종", "종목코드", "종목명", "현재가"]
df1 = DataFrame(data=data, columns=columns)

# 두 번째 데이터프레임
data = [
    ["은행", 2.92],
    ["보험", 0.37],
    ["화학", 0.06],
    ["전기전자", -2.43]
]

columns = ["항목", "등락률"]
df2 = DataFrame(data=data, columns=columns)

'''
columns 에서 원래 "업종"으로 되어 있었는데 "항목"으로 임의로 바꿔봄
그러니까 merge 시키면 에러 발생
key 매칭이 안 되니까
'''
#df=pd.merge(left=df1,right=df2,how='outer',on='업종') #에러 발생
#df

#근데 left_on right_on 이런거 하지 말고, 그냥 col의 이름명을 통일 시켜주는 것이 제일 편해요ㅎㅎ

df=pd.merge(left=df1,right=df2,how='outer',left_on='업종', right_on="항목") #이렇게 해결 가능
df
```

```python
# 첫 번째 데이터프레임
data = [
    ["전기전자", "005930", "삼성전자", 74400],
    ["화학", "051910", "LG화학", 896000],
    ["서비스업", "035720", "카카오", 101500]
]

columns = ["업종", "종목코드", "종목명", "현재가"]
df1 = DataFrame(data=data, columns=columns)


df1=df1.set_index("업종")
print(df1) #이제부터 업종을 index로서 사용하겠다



# 두 번째 데이터프레임
data = [
    ["은행", 2.92],
    ["보험", 0.37],
    ["화학", 0.06],
    ["전기전자", -2.43]
]

columns = ["항목", "등락률"]
df2 = DataFrame(data=data, columns=columns)

df2=df2.set_index("항목")
print(df2) #이제부터 항목을 index로서 사용하겠다
```

```python
#merge는 기본 값이 inner join이었음
#근데 join은 기본 값이 left join임

'''
강사님이 추천하는 방식은 how를 작성해서 명시해주는 것이 차라리 마음이 편함

예시
df1.join(other=df2, how ="inner")
df1.join(other=df2, how ="left")
어차피 left/right/outer/join 뭐 4가지 정도밖에 없음
'''

df1.join(other=df2) #join의 기본 디폴트는 left 
```

##연도 별로 그룹바이 연산, 시가총액 평균


```python
# 그룹화 및 평균 계산
result = df.groupby('연도')['시가총액'].mean().to_frame()
result.columns=['시가총액평균']

print(result)
```

```python
df=df.join(result,on="연도")
df

#on을 쓰게 되면 해당 "연도"가 index 이든지 col이든지 상관없이 같으 것을 찾고 그것을 기준으로 정렬해줌
# 근데 가장 좋은 건 그냥 하나로 형태를 일치 시켜주는 것이 제일 좋아요!
```

```python
# 시가총액평균을 기준값으로 
# 새로운 컬럼 '규모'라고 만들어주시고, 
# 안에 들어가는 값은 대형주, 중소형주로 구분해주세요
# numpy.where을 써서 만들어주세요

# 2017년의 경우 시가총액평균이 333이니까 이거보다 크면 대형주이고 작으면 중소형주임

```

```python
import numpy as np
#numpy.where을 활용해서 series 연산을 진행한 것입니다

df['규모']=np.where(df['시가총액']>=df['시가총액평균'], "대형주", "중소형주")
df
```
