```python
import pandas as pd
from scipy import stats

'''
chat gpt의 코드
'''


# 문제 1: 표본평균 구하기 (Height)
sample_mean = round(data['Height'].mean(), 2)
print("표본 평균 (X):", sample_mean)

# 문제 2: 검정통계량 구하기
population_mean = 75  # 알려진 모평균
sample_std = data['Height'].std(ddof=1)  # 표본 표준편차
sample_size = len(data['Height'])  # 표본 크기

# t-검정 통계량 계산
t_statistic = (sample_mean - population_mean) / (sample_std / (sample_size ** 0.5))
t_statistic_rounded = round(t_statistic, 2)
print("검정통계량 (t):", t_statistic_rounded)

# 문제 3: p-값 구하기
p_value = stats.t.sf(abs(t_statistic), df=sample_size-1) * 2
p_value_rounded = round(p_value, 4)
print("p-값:", p_value_rounded)

# 가설검정 결과 판단
alpha = 0.05
if p_value_rounded < alpha:
    print("결과: 귀무가설 기각")
else:
    print("결과: 귀무가설 채택")
```
