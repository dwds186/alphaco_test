```python
#graph_object

fig=go.Figure(data=[go.Bar(x=["a", "b", "c"], y = [1,3,2])],
              layout=go.Layout(title=go.layout.Title(text="Title 설정"))
             
             )
fig.show()
```

```python
#graph_object

fig=go.Figure(data=[go.Bar(x=["a", "b", "c"], y = [1,3,2])])

fig.update_layout(width=600, height=300)

fig.show()
```

graph_object보다는 express가 조금 더 간단함

```python
#graph_object

fig=go.Figure(data=[go.Bar(x=["a", "b", "c"], y = [1,3,2])])

fig.update_layout(width=600,
                  height=300,
                  margin_l=200, #left
                  margin_r=30, #right
                  #background color
                  paper_bgcolor="lightblue"
                 )
#update_layout은 전부 그래프의 외관과 관련 사항을 컨트롤 합니다!
fig.show()
```

```python
#공식 사이트에 있는 샘플 코드 예시

import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['AAPL.Open'],
                high=df['AAPL.High'],
                low=df['AAPL.Low'],
                close=df['AAPL.Close'])])

fig.show()
```

```python
import yfinance as yf
import plotly.graph_objects as go
#인도네시아 에너지ㅋㅋㅋ
indo = yf.download("INDO", start="2023-01-01", end = "2024-09-30")
indo.head()
indo.tail()

fig = go.Figure(data=[go.Candlestick(x=indo.index,
                open=indo['Open'],
                high=indo['High'],
                low=indo['Low'],
                close=indo['Close'])])

#update_layout
#update_traces
#둘 중에 어떤 것으로 세부사항을 컨트롤 할 수 있는지 chat gpt한테 물어보고 확인해서 직접 업데이트 해보세요!

#chat gpt한테 물어보니 차트 제목, 축 레이블은  layout에서 컨트롤 할 수 있다고 함

#update_layout
# 차트 제목, 축 레이블을 설정해보세요


fig.update_layout(
    title='인도네시아 에너지 차트(2023-2024)',
    xaxis_title='날짜',
    yaxis_title='주식 가격(USD)',
    hovermode='x',
    xaxis_rangeslider_visible=True,
    xaxis_rangeslider_thickness=0.5,
    paper_bgcolor='pink',
    plot_bgcolor='white'
)





#update_traces
# 캔들 차트의 색상을 변경해보세요 
fig.update_traces(
    increasing_line_color='red',
    decreasing_line_color='blue',
)


fig.show()
```
