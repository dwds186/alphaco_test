```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = final_data_1111

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# 6x2 subplot 생성
fig, axes = plt.subplots(6, 2, figsize=(15, 30))  # 6행, 2열

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # Filtered DataFrame (subset)
        subset = df[(df['성별'] == gender) & (df['연령대'] == age)]

        if not subset.empty:
            # '가맹점업종명_범주화'의 카운트 계산
            category_counts = subset['가맹점업종명_범주화_중분류'].value_counts().reset_index()
            category_counts.columns = ['가맹점업종명_범주화_중분류', '카운트']

            # 막대 그래프 생성
            ax = axes[plot_idx // 2, plot_idx % 2]
            sns.barplot(data=category_counts, x='가맹점업종명_범주화_중분류', y='카운트', ax=ax)
            ax.set_title(f"{gender} - {age} [가맹점업종명_범주화 카운트]")
            ax.set_xlabel("가맹점업종명_범주화")
            ax.set_ylabel("카운트")
            ax.tick_params(axis='x', rotation=45)

            plot_idx += 1

# 레이아웃 조정
plt.tight_layout()
plt.savefig('mid_category_counts.png')
plt.show()
```

```python
# 가맹점_광역시도별 승인 금액 합계 계산
regional_spending = final_data_1111.groupby(['성별', '연령대', '가맹점_광역시도'])['승인금액'].sum().reset_index()

# 히트맵 생성
regional_pivot = regional_spending.pivot_table(index='연령대', columns='가맹점_광역시도', values='승인금액', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(regional_pivot, annot=False, cmap='viridis', linewidths=0.5)
plt.title("성별 및 연령대별 지역 소비 비교")
plt.xlabel("가맹점 광역시도")
plt.ylabel("연령대")
plt.tight_layout()
plt.show()
```

```python
# 모든 업종의 데이터를 사용하여 소비 분포 계산
industry_distribution = final_data_1111.groupby(['성별', '연령대', '가맹점업종명_범주화_중분류'])['승인금액'].sum().reset_index()

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=industry_distribution, x='연령대', y='승인금액', hue='가맹점업종명_범주화_중분류', palette='viridis')
plt.title("성별 및 연령대별 모든 업종 소비 비율")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점 업종", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 가맹점_시군구별 승인 금액 합계 계산
district_spending = final_data_1111.groupby(['성별', '연령대', '가맹점_시군구'])['승인금액'].sum().reset_index()

# 상위 10개 시군구 필터링
top_districts = district_spending.groupby('가맹점_시군구')['승인금액'].sum().nlargest(10).index
district_spending = district_spending[district_spending['가맹점_시군구'].isin(top_districts)]

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=district_spending, x='연령대', y='승인금액', hue='가맹점_시군구', palette='viridis')
plt.title("성별 및 연령대별 주요 가맹점_시군구 소비 집중도")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점_시군구")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = final_data_1111_noInternet

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# 6x2 subplot 생성
fig, axes = plt.subplots(6, 2, figsize=(15, 30))  # 6행, 2열

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # Filtered DataFrame (subset)
        subset = df[(df['성별'] == gender) & (df['연령대'] == age)]

        if not subset.empty:
            # '가맹점업종명_범주화'의 카운트 계산
            category_counts = subset['가맹점업종명_범주화_중분류'].value_counts().reset_index()
            category_counts.columns = ['가맹점업종명_범주화_중분류', '카운트']

            # 막대 그래프 생성
            ax = axes[plot_idx // 2, plot_idx % 2]
            sns.barplot(data=category_counts, x='가맹점업종명_범주화_중분류', y='카운트', ax=ax)
            ax.set_title(f"{gender} - {age} [가맹점업종명_범주화 카운트]")
            ax.set_xlabel("가맹점업종명_범주화")
            ax.set_ylabel("카운트")
            ax.tick_params(axis='x', rotation=45)

            plot_idx += 1

# 레이아웃 조정
plt.tight_layout()
plt.savefig('mid_category_counts_noInternet.png')
plt.show()
```

```python
# 모든 업종의 데이터를 사용하여 소비 분포 계산
industry_distribution = final_data_1111_noInternet.groupby(['성별', '연령대', '가맹점업종명_범주화_중분류'])['승인금액'].sum().reset_index()

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=industry_distribution, x='연령대', y='승인금액', hue='가맹점업종명_범주화_중분류', palette='viridis')
plt.title("성별 및 연령대별 모든 업종 소비 비율")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점 업종", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 가맹점_시군구별 승인 금액 합계 계산
district_spending = final_data_1111_noInternet.groupby(['성별', '연령대', '가맹점_시군구'])['승인금액'].sum().reset_index()

# 상위 10개 시군구 필터링
top_districts = district_spending.groupby('가맹점_시군구')['승인금액'].sum().nlargest(10).index
district_spending = district_spending[district_spending['가맹점_시군구'].isin(top_districts)]

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=district_spending, x='연령대', y='승인금액', hue='가맹점_시군구', palette='viridis')
plt.title("성별 및 연령대별 주요 가맹점_시군구 소비 집중도")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점_시군구")
plt.tight_layout()
plt.show()
```
