```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv("final_data_1108.csv")  # 데이터 불러오기 예시

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# 6x2 subplot 생성
fig, axes = plt.subplots(6, 2, figsize=(15, 30))  # 6행, 2열

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # Filtered DataFrame (subset)
        subset = df[(df['성별'] == gender) & (df['연령대'] == age)]

        if not subset.empty:
            # '가맹점업종명_범주화'의 카운트 계산
            category_counts = subset['가맹점업종명_범주화'].value_counts().reset_index()
            category_counts.columns = ['가맹점업종명_범주화', '카운트']

            # 막대 그래프 생성
            ax = axes[plot_idx // 2, plot_idx % 2]
            sns.barplot(data=category_counts, x='가맹점업종명_범주화', y='카운트', ax=ax)
            ax.set_title(f"{gender} - {age} [가맹점업종명_범주화 카운트]")
            ax.set_xlabel("가맹점업종명_범주화")
            ax.set_ylabel("카운트")
            ax.tick_params(axis='x', rotation=45)

            plot_idx += 1

# 레이아웃 조정
plt.tight_layout()
plt.savefig('category_counts.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv("final_data_1108.csv")  # 데이터 불러오기 예시

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# Subplot 설정 (2열 6행)
fig, axes = plt.subplots(6, 2, figsize=(14, 24))  # 6 rows, 2 columns

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # 각 성별-연령대 그룹에 대해 서울특별시와 경기도의 행 개수를 인구 수로 간주
        subset = df[(df['성별'] == gender) & (df['연령대'] == age) & df['가맹점_광역시도'].isin(['서울특별시', '경기도'])]
        
        # '서울특별시'와 '경기도'의 행 개수 (각각 인구수로 간주)
        population_seoul = subset[subset['가맹점_광역시도'] == '서울특별시'].shape[0]
        population_gyeonggi = subset[subset['가맹점_광역시도'] == '경기도'].shape[0]
        
        # 서울과 경기도 인구 데이터를 위한 데이터프레임 생성
        population_data = pd.DataFrame({
            '가맹점_광역시도': ['서울특별시', '경기도'],
            '인구수': [population_seoul, population_gyeonggi]
        })

        # 현재 subplot에 막대 그래프 생성
        ax = axes[plot_idx // 2, plot_idx % 2]
        sns.barplot(data=population_data, x='가맹점_광역시도', y='인구수', hue='가맹점_광역시도', ax=ax, palette='viridis', legend=False)
        ax.set_title(f"{gender} - {age} 가맹점_광역시도 [서울특별시 vs 경기도]")
        ax.set_xlabel("Region")
        ax.set_ylabel("sum of every industry")
        
        plot_idx += 1

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('가맹점_광역시도 서울 vs 경기.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv("final_data_1108.csv")  # 데이터 불러오기 예시

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# Subplot 설정 (2열 6행)
fig, axes = plt.subplots(6, 2, figsize=(14, 24))  # 6 rows, 2 columns

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # 각 성별-연령대 그룹에 대해 고객등급 비율 계산
        subset = df[(df['성별'] == gender) & (df['연령대'] == age)]
        
        # 고객등급의 비율 계산
        grade_counts = subset['고객등급'].value_counts(normalize=True) * 100
        grade_data = pd.DataFrame({
            '고객등급': grade_counts.index,
            '비율': grade_counts.values
        })

        # 현재 subplot에 막대 그래프 생성
        ax = axes[plot_idx // 2, plot_idx % 2]
        sns.barplot(data=grade_data, x='고객등급', y='비율', hue='고객등급', ax=ax, palette='viridis', legend=False)
        ax.set_title(f"{gender} - {age} [고객등급 비율]")
        ax.set_xlabel("Customer Grade")
        ax.set_ylabel("Percentage (%)")
        
        plot_idx += 1

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('customer_grade_ratio_subplots.png')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv("final_data_1108.csv")  # 데이터 불러오기 예시

# 연령대 그룹 정의
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# Subplot 설정 (2열 6행)
fig, axes = plt.subplots(6, 2, figsize=(14, 24))  # 6 rows, 2 columns

# Plot index
plot_idx = 0
for age in ages:
    # 해당 연령대에 대한 데이터 필터링
    subset = df[df['연령대'] == age]
    
    # 성별별 행 개수 계산 (남자와 여자의 수)
    gender_counts = subset['성별'].value_counts().reset_index()
    gender_counts.columns = ['성별', '숫자']

    # 현재 subplot에 막대 그래프 생성
    ax = axes[plot_idx // 2, plot_idx % 2]
    sns.barplot(data=gender_counts, x='성별', y='숫자', hue='성별', ax=ax, palette='viridis', legend=False)
    ax.set_title(f"{age} 남자 vs 여자 인구 수")
    ax.set_xlabel("Gender")
    ax.set_ylabel("Count")
    
    plot_idx += 1

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('gender_comparison_by_age_subplots.png')
plt.show()
```
