```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 폰트 설정
plt.rc('font', family='Malgun Gothic')  # Windows의 한글 폰트 설정

# 음수 표시 설정 (깨지는 문제 방지)
plt.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd
adjusted_plot_data = pd.read_csv('adjusted_plot_data.csv')


# 데이터 시각화 코드
fig, ax = plt.subplots(figsize=(10, 8))

# Scatter plot with adjusted Korean font
for card_type, group_data in adjusted_plot_data.groupby('카드구분'):
    ax.scatter(
        group_data['x_score'],
        group_data['y_score'],
        label=card_type,
        alpha=0.7,
        s=100
    )

for _, row in adjusted_plot_data.iterrows():
    ax.text(row['x_score'], row['y_score'], row['카드명'], fontsize=8, ha='right')

# 그래프 라벨 및 타이틀
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.set_xlabel("금리/수수료 혜택 → 리워드 혜택", fontsize=12)
ax.set_ylabel("프리미엄 카드 → 생활 밀착형 카드", fontsize=12)
ax.set_title("iM뱅크 카드 포지셔닝 맵 (조정된 분포)", fontsize=16)
ax.legend(title="카드 구분")

plt.grid(alpha=0.3)
plt.show()
```

1. 타겟 고객군 정의
MZ세대 (20~40대):

디지털 네이티브, 모바일 중심의 소비 패턴
친환경, 윤리 소비 등 가치 중심 소비 추구
여행, 외식, 여가활동에 대한 관심 증가
액티브 시니어 (50대 이상):

자산이 높은 중장년층
건강 관리, 웰니스, 프리미엄 서비스에 높은 관심
2. 혜택 아이디어
(1) MZ 세대 타겟
구독 서비스 특화 카드:

주요 스트리밍 플랫폼(넷플릭스, 유튜브 프리미엄, 스포티파이) 및 배달앱 할인
구독 서비스 이용 금액의 일정 비율 캐시백 (10~20%)
ESG 및 친환경 혜택:

대중교통(지하철, 버스) 이용 시 포인트 적립
친환경 제품 구매 시 추가 할인(예: 텀블러 사용 시 카페 할인)
디지털 결제 보너스:

간편결제(삼성페이, 카카오페이, 애플페이) 사용 시 수수료 면제 또는 캐시백
모바일 앱 기반의 전용 리워드 프로그램 (게임화 요소 포함)
짧은 사용 주기의 할인/혜택:

월 단위로 혜택을 업데이트하여 새로운 소비 트렌드에 맞춘 혜택 제공
예: 특정 이커머스 플랫폼에서 한정 할인
(2) 액티브 시니어 타겟
건강 관리와 웰니스:

헬스케어 및 건강기기(병원, 약국, 건강보조식품) 구매 시 포인트 적립
건강검진 센터 이용료 할인
웰니스 여행(온천, 리조트) 패키지 할인
로컬 경제 활성화 카드:

지역 상점, 전통시장, 로컬 음식점에서 사용 시 추가 혜택
지역 특산품 구매 시 캐시백 제공
프리미엄 혜택:

공항 라운지 무료 이용
호텔 숙박 할인 및 고급 레스토랑 예약 서비스
금융 관리 지원:

자산 관리 프로그램(금융 컨설팅 서비스) 무료 제공
고액 자산 관리 고객을 위한 맞춤형 리워드 (펀드/주식 투자 혜택)
3. 기술적 차별화 요소
데이터 기반 맞춤형 혜택:

AI 분석을 통해 사용자의 소비 패턴에 따라 혜택 자동 추천
고객 선호도를 반영한 맞춤형 리워드 제공
디지털 채널 전용 카드:

모바일 앱에서만 발급 및 관리 가능
카드 발급 과정의 완전한 디지털화
탄소 배출 감축 인증:

카드 이용 시 탄소 배출량을 줄이는 활동을 인증하는 프로그램
4. 포지셔닝
X축: 새로운 디지털 서비스 중심의 리워드 혜택
Y축: 가치 중심(친환경, 웰니스) 또는 초프리미엄 서비스
