```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

final_plot_data_reprocessed = pd.read_csv('final_plot_data_reprocessed.csv')
data = final_plot_data_reprocessed

# 데이터 전처리
data = data.fillna('').applymap(lambda x: str(x) if not pd.isna(x) else "")  # 모든 값을 문자열로 변환

# 연회비 정리 함수
def parse_annual_fee(value):
    try:
        fees = [int(fee.replace(',', '').strip()) for fee in value.split() if fee.replace(',', '').isdigit()]
        return max(fees) if fees else 0
    except:
        return 0

data['연회비_정리'] = data['연회비'].apply(parse_annual_fee)

# 신용카드/체크카드 구분
data['카드구분'] = data['댜'].apply(lambda x: '신용카드' if '신용' in x else '체크카드')

# 스코어링 함수
def calculate_y_score(row):
    premium_keywords = ['프리미엄', 'VIP', '특급', '라운지', '고급']
    premium_score = sum(any(kw in str(row[col]) for kw in premium_keywords) for col in data.columns)
    return (premium_score * 15) + (np.log1p(float(row['연회비_정리'])) * 10)

def calculate_x_score(row):
    reward_keywords = ['캐시백', '포인트', '마일리지', '리워드']
    fee_keywords = ['수수료', '금리', '이자', '할부']
    reward_score = sum(any(kw in str(row[col]) for kw in reward_keywords) for col in data.columns)
    fee_score = sum(any(kw in str(row[col]) for kw in fee_keywords) for col in data.columns)
    return (fee_score * 10) - (reward_score * 10)

# 스코어 계산
data['y_score'] = data.apply(calculate_y_score, axis=1)
data['x_score'] = data.apply(calculate_x_score, axis=1)

# 필터링
final_plot_data = data[['카드명', '카드구분', 'y_score', 'x_score']].dropna()

# 시각화
fig, ax = plt.subplots(figsize=(15, 12))
for card_type, group_data in final_plot_data.groupby('카드구분'):
    ax.scatter(
        group_data['x_score'].astype(float),
        group_data['y_score'].astype(float),
        label=card_type,
        alpha=0.7,
        s=150
    )

# 카드 이름 표시
for _, row in final_plot_data.iterrows():
    ax.text(float(row['x_score']), float(row['y_score']), row['카드명'], fontsize=9, ha='right', va='center')

# 그래프 구성
ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
ax.set_xlabel("금리/수수료 혜택 → 리워드 혜택", fontsize=14)
ax.set_ylabel("프리미엄 카드 → 생활 밀착형 카드", fontsize=14)
ax.set_title("iM뱅크 카드 포지셔닝 맵 (수정 완료)", fontsize=18)
ax.legend(title="카드 구분", fontsize=12)

plt.grid(alpha=0.3)
plt.show()

# 결과 저장
output_file = "final_plot_data_reprocessed.csv"
final_plot_data.to_csv(output_file, index=False)
print(f"Processed data saved to {output_file}")
```
