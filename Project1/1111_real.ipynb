```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 예시 (사용자의 실제 데이터로 대체)
data = pd.read_csv('final_data_1111.csv') # 실제 데이터 경로 사용

# "서울특별시"와 "경기도" 데이터를 필터링
seoul_data = data[data['가맹점_광역시도'] == "서울특별시"]
gyeonggi_data = data[data['가맹점_광역시도'] == "경기도"]

# '가맹점업종명_범주화' 빈도 계산
seoul_categories = seoul_data['가맹점업종명_범주화'].value_counts()
gyeonggi_categories = gyeonggi_data['가맹점업종명_범주화'].value_counts()

# 두 지역의 데이터프레임으로 정리
comparison_df = pd.DataFrame({
    '서울특별시': seoul_categories,
    '경기도': gyeonggi_categories
}).fillna(0)

# 시각화
comparison_df.plot(kind='bar', figsize=(14, 8))
plt.title('서울특별시 vs 경기도 - 가맹점업종명_범주화 빈도 비교')
plt.xlabel('가맹점업종명_범주화')
plt.ylabel('빈도수')
plt.xticks(rotation=45)
plt.legend(title='지역')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = final_data_1111_noInternet

# 성별 및 연령대 그룹 정의
genders = df['성별'].unique()  # 예: ['남', '여']
ages = df['연령대'].unique()  # 예: [20, 30, 40, 50, 60, 70]

# 6x2 subplot 생성
fig, axes = plt.subplots(6, 2, figsize=(15, 30))  # 6행, 2열

# Plot index
plot_idx = 0
for age in ages:
    for gender in genders:
        # Filtered DataFrame (subset)
        subset = df[(df['성별'] == gender) & (df['연령대'] == age)]

        if not subset.empty:
            # '가맹점업종명_범주화'의 카운트 계산
            category_counts = subset['가맹점업종명_범주화'].value_counts().reset_index()
            category_counts.columns = ['가맹점업종명_범주화', '카운트']

            # 막대 그래프 생성
            ax = axes[plot_idx // 2, plot_idx % 2]
            sns.barplot(data=category_counts, x='가맹점업종명_범주화', y='카운트', ax=ax)
            ax.set_title(f"{gender} - {age} [가맹점업종명_범주화 카운트]")
            ax.set_xlabel("가맹점업종명_범주화")
            ax.set_ylabel("카운트")
            ax.tick_params(axis='x', rotation=45)

            plot_idx += 1

# 레이아웃 조정
plt.tight_layout()
plt.savefig('category_counts_noInternet_noInternet.png')
plt.show()
```

```python
# 가맹점_광역시도별 승인 금액 합계 계산
regional_spending = final_data_1111_noInternet.groupby(['성별', '연령대', '가맹점_광역시도'])['승인금액'].sum().reset_index()

# 히트맵 생성
regional_pivot = regional_spending.pivot_table(index='연령대', columns='가맹점_광역시도', values='승인금액', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(regional_pivot, annot=False, cmap='viridis', linewidths=0.5)
plt.title("성별 및 연령대별 지역 소비 비교")
plt.xlabel("가맹점 광역시도")
plt.ylabel("연령대")
plt.tight_layout()
plt.show()
```

```python
# 주요 업종 필터링
top_industries = final_data_1111_noInternet.groupby('가맹점업종명')['승인금액'].sum().nlargest(5).index
industry_spending = final_data_1111_noInternet[final_data_1111_noInternet['가맹점업종명'].isin(top_industries)]
industry_distribution = industry_spending.groupby(['성별', '연령대', '가맹점업종명'])['승인금액'].sum().reset_index()

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=industry_distribution, x='연령대', y='승인금액', hue='가맹점업종명', palette='viridis')
plt.title("성별 및 연령대별 주요 업종 소비 비율")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점 업종")
plt.tight_layout()
plt.show()
```

```python
# 가맹점_시군구별 승인 금액 합계 계산
district_spending = final_data_1111_noInternet.groupby(['성별', '연령대', '가맹점_시군구'])['승인금액'].sum().reset_index()

# 상위 10개 시군구 필터링
top_districts = district_spending.groupby('가맹점_시군구')['승인금액'].sum().nlargest(10).index
district_spending = district_spending[district_spending['가맹점_시군구'].isin(top_districts)]

# 스택형 막대 그래프 생성
plt.figure(figsize=(14, 8))
sns.barplot(data=district_spending, x='연령대', y='승인금액', hue='가맹점_시군구', palette='viridis')
plt.title("성별 및 연령대별 주요 가맹점_시군구 소비 집중도")
plt.xlabel("연령대")
plt.ylabel("승인 금액 합계")
plt.legend(title="가맹점_시군구")
plt.tight_layout()
plt.show()
```

# 기타

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = final_data_1111 # 데이터 불러오기 예시

# "고객ID"를 기준으로 중복 제거하여 순수한 고객 데이터 생성
unique_customers = df[['고객ID', '성별', '연령대']].drop_duplicates()

# 연령대 그룹 정의
ages = unique_customers['연령대'].unique()

# Subplot 설정 (2열 6행)
fig, axes = plt.subplots(6, 2, figsize=(14, 24))  # 6 rows, 2 columns

# Plot index
plot_idx = 0
for age in sorted(ages):
    # 해당 연령대에 대한 데이터 필터링
    subset = unique_customers[unique_customers['연령대'] == age]
    
    # 성별별 행 개수 계산 (남자와 여자의 수)
    gender_counts = subset['성별'].value_counts().reset_index()
    gender_counts.columns = ['성별', '숫자']

    # 현재 subplot에 막대 그래프 생성
    ax = axes[plot_idx // 2, plot_idx % 2]
    sns.barplot(data=gender_counts, x='성별', y='숫자', ax=ax, palette='viridis', dodge=False)
    ax.set_title(f"{age}대 남자 vs 여자 순수 인구 수")
    ax.set_xlabel("성별")
    ax.set_ylabel("수")
    
    # 범례가 None일 경우를 대비
    if ax.get_legend() is not None:
        ax.get_legend().remove()  # Remove the legend for each subplot
    
    plot_idx += 1

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.savefig('gender_comparison_by_age_subplots_fixed_sample.png')
plt.show()

```
